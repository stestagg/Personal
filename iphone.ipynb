# Looking at the 'iphone slow' search data

```python
import csv
import datetime
%pylab inline
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
iphone_data = list(csv.DictReader(open("iphone.csv", "r")))
```

```python
iphone_data
```

```python
iphone_slow = list(csv.DictReader(open("slow.csv", "r")))
```

```python
iphone_slow
```

```python
dl_slow = list(csv.DictReader(open("dlslow.csv", "r")))
```

```python
dl_slow
```

```python
app_slow = list(csv.DictReader(open("appslow.csv", "r")))
```

```python
app_slow
```

```python
[a for a in zip(iphone_data, iphone_slow, dl_slow, app_slow) if len(set([x["From"] for x in a])) != 1]
```

```python
combined = [(
             datetime.datetime.strptime(s["From"], "%Y-%m-%d").date(), 
             int(i["iphone"]), 
             int(s["iphone slow"]), 
             int(dl["iphone download slow"]), 
             int(app["iphone apps slow"])
             ) for i,s,dl,app in zip(iphone_data, iphone_slow, dl_slow, app_slow)]
```

```python
combined_stripped = [a for a in combined if any([x!=0 for x in a[1:]])]
```

```python
combined_stripped
```

```python
relative_data = [(d, i-s, i-dl, i-a) for d, i, s, dl, a  in combined_stripped]
```

```python
I_DATE = 0
I_SLOW = 1
I_DL_SLOW = 2
I_APP_SLOW = 3
```

```python
release_dates = [datetime.date(2007, 6, 29), 
                 datetime.date(2008, 7, 11), 
                 datetime.date(2009, 7, 19), 
                 datetime.date(2010, 6, 24),
                 datetime.date(2011, 10, 14),
                 datetime.date(2012, 9, 21),
                 datetime.date(2013, 9, 20),
                 ]
```

## Relative searches for 'iphone slow'

```python
plt.plot([d[I_DATE] for d in relative_data], [d[I_SLOW] for d in relative_data])
plt.vlines(release_dates, [-100] * len(release_dates), [100] * len(release_dates))
```

## Relative searches for 'iphone apps slow'

```python
plt.plot([d[I_DATE] for d in relative_data], [d[I_APP_SLOW] for d in relative_data], color="red")
plt.vlines(release_dates, [-100] * len(release_dates), [100] * len(release_dates))
```

## Relative searches for 'iphone download slow'

```python
plt.plot([d[I_DATE] for d in relative_data], [d[I_DL_SLOW] for d in relative_data], color="green")
plt.vlines(release_dates, [-100] * len(release_dates), [100] * len(release_dates))
```
